## Imports

```python
import googlemaps
import pandas as pd
import numpy as np
import re
import os
import folium
```

## Read-in

```python
df = pd.read_csv("ChicagoPortfolioMap_04_26_24 - Sheet1.csv")
```

## Google Maps Geolocater

```python
%store -r google_maps_API_Key
gmaps_key = googlemaps.Client(key=google_maps_API_Key)
```

```python
df['full_address'] = df['full_address'] + ' Chicago'
```

```python
def geocode(add):
    g = gmaps_key.geocode(add)
    lat = g[0]["geometry"]["location"]["lat"]
    lng = g[0]["geometry"]["location"]["lng"]
    return (lat, lng)

df['geocoded'] = df['full_address'].apply(geocode)

df['geocoded'] = df['geocoded'].astype(str)
df[['lat', 'lon']] = df['geocoded'].str.strip('()').str.split(', ', expand=True)
df['lat'] = df['lat'].astype(float)
df['lon'] = df['lon'].astype(float)
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.columns
```

```python
def popup_html(row):
    portfolio_name = row['portfolio_name']
    address = row['full_address']
    
    html = '''<!DOCTYPE html>
    <html>
    <strong>Building Name: </strong>{}'''.format(portfolio_name) + '''<br>
    <strong>Address: </strong>{}'''.format(address) + '''<br>
    </html>
    '''
    return html
```

```python
df['color'].value_counts()
```

```python
from folium.plugins import MarkerCluster

title_html = '''
              <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(f'Cook County Portfolio Map')

### Create map container ###
m = folium.Map(location=df[["lat", "lon"]].mean().to_list(),zoom_start=11.5,tiles='OpenStreetmap')


for index, row in df.iterrows():
    marker = folium.Marker(
        location=[row['lat'], row['lon']],
        radius=5,
        fill=True,
        icon=folium.Icon(color=row['color']),
        popup=folium.Popup(popup_html(row), max_width=400))
    marker.add_to(m)
            
# Display map
m
```

```python
m.save('index.html')
```
