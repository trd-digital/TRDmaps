```python
df_top_ten_final = pd.read_csv('df_top_ten_final.csv')
```

```python
df_top_ten_final[' bldgarea '] = df_top_ten_final[' bldgarea '].str.replace(',','')
df_top_ten_final[' bldgarea '] = df_top_ten_final[' bldgarea '].str.replace('-','')
df_top_ten_final[' bldgarea '] = df_top_ten_final[' bldgarea '].str.replace('     ','')
```

```python
df_top_ten_final[' bldgarea '] = pd.to_numeric(df_top_ten_final[' bldgarea '])
```

```python
df_top_ten_final[' bldgarea '] = df_top_ten_final[' bldgarea '].apply(lambda x : "{:,}".format(x))
```

```python
df_NYU = df_top_ten_final.loc[df_top_ten_final['Final'] == 'NEW YORK UNIVERSITY']
df_CU = df_top_ten_final.loc[df_top_ten_final['Final'] == 'COLUMBIA UNIVERSITY']
df_Corn = df_top_ten_final.loc[df_top_ten_final['Final'] == 'CORNELL UNIVERSITY']
df_SJU = df_top_ten_final.loc[df_top_ten_final['Final'] == "ST JOHN'S UNIVERSITY"]
df_Ford = df_top_ten_final.loc[df_top_ten_final['Final'] == 'FORDHAM UNIVERSITY']
```

```python
df_NYU.columns
```

```python
m = folium.Map(location=df_top_ten_final[['latitude','longitude']].mean().to_list(), zoom_start=11, tiles='CartoDB positron')

title_html = '''
              <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(f'NYC College Footprint')

# marker_clusterNYU = MarkerCluster().add_to(m)

for i,r in df_NYU.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r[' bldgarea '],r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[1]}{bold_end}{new_line}\
    Address: {bold_start}{info[2]}{bold_end}{new_line}\
    SqFt (this buidling): {bold_start}{info[0]}{bold_end}{new_line}\
    Total NYU SqFt: {bold_start}14,030,662{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='darkpurple'),
                 tooltip=text)\
    .add_to(m)
    
# marker_clusterCU = MarkerCluster().add_to(m)

for i,r in df_CU.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r[' bldgarea '],r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[1]}{bold_end}{new_line}\
    Address: {bold_start}{info[2]}{bold_end}{new_line}\
    SqFt (this buidling): {bold_start}{info[0]}{bold_end}{new_line}\
    Total Columbia SqFt: {bold_start}13,950,318{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='lightblue'),
                 tooltip=text)\
    .add_to(m)
    
# marker_clusterSJU = MarkerCluster().add_to(m)
    
for i,r in df_SJU.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r[' bldgarea '],r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[1]}{bold_end}{new_line}\
    Address: {bold_start}{info[2]}{bold_end}{new_line}\
    SqFt (this buidling): {bold_start}{info[0]}{bold_end}{new_line}\
    Total SJU SqFt: {bold_start}5,091,959{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='darkred'),
                 tooltip=text)\
    .add_to(m)
    
# marker_clusterCorn = MarkerCluster().add_to(m)
    
for i,r in df_Corn.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r[' bldgarea '],r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[1]}{bold_end}{new_line}\
    Address: {bold_start}{info[2]}{bold_end}{new_line}\
    SqFt (this buidling): {bold_start}{info[0]}{bold_end}{new_line}\
    Total Cornell SqFt: {bold_start}5,469,412{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='red'),
                 tooltip=text)\
    .add_to(m)
    
# marker_clusterFord = MarkerCluster().add_to(m)
    
for i,r in df_Ford.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r[' bldgarea '],r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[1]}{bold_end}{new_line}\
    Address: {bold_start}{info[2]}{bold_end}{new_line}\
    SqFt (this buidling): {bold_start}{info[0]}{bold_end}{new_line}\
    Total Fordham SqFt: {bold_start}3,093,927{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='lightred'),
                 tooltip=text)\
    .add_to(m)

m.get_root().html.add_child(folium.Element(title_html))    

m
```
