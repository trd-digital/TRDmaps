```python
df_top_ten_final = pd.read_csv('df_top_ten_final.csv')
```

```python
m = folium.Map(location=df_top_ten_final[['latitude','longitude']].mean().to_list(), zoom_start=10, tiles='CartoDB positron')

title_html = '''
              <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(f'NYC College Footprint')

marker_cluster = MarkerCluster().add_to(m)

for i,r in df_top_ten_final.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[0]}{bold_end}{new_line}\
    Address: {bold_start}{info[1]}{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='orange'),
                 tooltip=text)\
    .add_to(marker_cluster)
```

```python
m.save('index.html')
m
```
