## Imports and Set Options

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import os

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 5000)
```

## Readin and analysis

```python
df = pd.read_csv('University_College Footprint 0922 - All Data.csv')
```

```python
df.columns
```

```python
top_ten_list = [
    "COLUMBIA UNIVERSITY",
    "NEW YORK UNIVERSITY",
    "ST JOHN'S UNIVERSITY",
    "FORDHAM UNIVERSITY",
    "ALBERT EINSTEIN COLLEGE OF MEDICINE",
    "THE ROCKEFELLER UNIVERSITY",
    "CORNELL UNIVERSITY",
    "PRATT INSTITUTE",
    "YESHIVA UNIVERSITY",
    "MANHATTAN COLLEGE",
    "BARNARD COLLEGE"
]
```

```python
df_top_ten = df.loc[df['Final'].isin(top_ten_list)]
```

```python
df_top_ten.Final.value_counts()
```

```python
df_top_ten[df_top_ten['latitude'].isna()]
```

```python
x = df_top_ten[df_top_ten['longitude'].isna()]
x.to_csv('college_no_lat_lon.csv')
```

```python
len(df_top_ten)
```

```python
df_top_ten = df_top_ten.dropna(subset=['latitude','longitude'])
```

```python
m = folium.Map(location=df_top_ten[['latitude','longitude']].mean().to_list(), zoom_start=10, tiles='CartoDB positron')

title_html = '''
              <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(f'NYC College Footprint')

marker_cluster = MarkerCluster().add_to(m)

for i,r in df_top_ten.iterrows():
    location = (r['latitude'],r['longitude'])
    info = (r['Final'],r['address'])
    info = list(info)
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    text = f'College: {bold_start}{info[0]}{bold_end}{new_line}\
    Address: {bold_start}{info[1]}{bold_end}'
    folium.Marker(location=location,icon=folium.Icon(color='orange'),
                 tooltip=text)\
    .add_to(marker_cluster)
```
