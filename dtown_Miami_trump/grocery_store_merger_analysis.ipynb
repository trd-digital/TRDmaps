### Required Input Fields

- input: (restaurant, 123 Main Street) A place name, address or category of establishments
- inputtype: Either "textquery" or "phonenumber"

### Optional Input Fields

- address_component
- adr_address
- business_status
- formatted_address
- geometry
- name

- locationbias: defaults to the user's IP address if no other parems are specified. Use the `circle` paremeter. It takes a lat/lon pair along with a radius specification in meters.
