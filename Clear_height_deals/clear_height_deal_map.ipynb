```python
from geopy.geocoders import GoogleV3
import pandas as pd
import numpy as np
import os
```

## Read-in

```python
df = pd.read_csv('clear_height_deals - Sheet1.csv')
```

```python
df['full_address'] = df['address'] + df['city']
```

## Google Maps API Geolocater Setup

```python
%store -r google_maps_API_Key
geolocator = GoogleV3(api_key=google_maps_API_Key)
```

```python
df['geo_address'] = df['full_address']
df['loc'] = df['geo_address'].apply(geolocator.geocode, timeout=10)
df['point'] = df['loc'].apply(lambda loc: tuple(loc.point) if loc else None)
df[['lat','lon','altitude']] = pd.DataFrame(df['point'].to_list(),index=df.index)
```

## Map Maker

```python
import folium
import branca

f = folium.Figure(width=750, height=750)
m = folium.Map(location=df[["lat", "lon"]].mean().to_list(),zoom_start=9)

title_html = '''
              <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(f"Clear Height's Industrial Purchases")

for i in range(0,len(df)):
    folium.Marker([df['lat'].iloc[i],df['lon'].iloc[i]]
                 ).add_to(m)

m.get_root().html.add_child(folium.Element(title_html))
m.fit_bounds(bounds=df[['lat','lon']].mean().to_list(),max_zoom=15)
# m.add_to(f)
m
```

```python
m.save('index.html')
```
